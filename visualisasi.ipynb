```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv")

#Hitung korelasi 
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriks Korelasi Data Regresi", fontsize=14)
plt.tight_layout()
plt.show()
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #73ad20e0; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;">  2. Boxplot Distribusi Variabel

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['Promo(X1)', 'Karyawan(X2)', 'Jual(Y)']], palette='Set2')
plt.title("Distribusi Variabel Promo, Karyawan, dan Penjualan", fontsize=14)
plt.tight_layout()
plt.show()
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #73ad20e0; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;">  3. Tabel Ringkasan Statistik

```python
import matplotlib.pyplot as plt

summary = df[['Promo(X1)', 'Karyawan(X2)', 'Jual(Y)']].describe().round(2)

#Metriks yang divisualisasikan
metrics = ['mean', 'std', 'min', '25%', '50%', '75%', 'max']

#Plot 
plt.figure(figsize=(10, 6))
summary.loc[metrics].T.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Ringkasan Statistik Variabel Promo, Karyawan, dan Penjualan', fontsize=14)
plt.ylabel('Nilai', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #2ea343ff; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;">Perbandingan Rata-Rata dari Kedua Dataset</div>

```python
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv")
url = "https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/refs/heads/main/kab_kota_sulsel.csv"
df2 = pd.read_csv(url, sep=';', header=0)

#Membersihkan kolom numerik di df2
for col in ['PAD', 'DAK', 'DAU', 'IPM']:
    df2[col] = df2[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df2[col] = pd.to_numeric(df2[col], errors='coerce')

#Menghitung rata-rata dari masing-masing DataFrame
avg_df1 = df1.select_dtypes(include='number').mean().to_frame(name='Rata-rata Data_Regresi')
avg_df2 = df2.select_dtypes(include='number').mean().to_frame(name='Rata-rata kab_kota_sulsel')

#Menggabungkan hasil rata-rata untuk perbandingan
comparison_df = pd.concat([avg_df1, avg_df2], axis=1)

print("\nPerbandingan Rata-rata dari Kedua Dataset:")
print(comparison_df.to_string())
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #73ad20e0; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;"> 1. Plot Batang </div>

```python
import matplotlib.pyplot as plt

#Plot batang horizontal
plt.figure(figsize=(10, 6))
comparison_df.plot(kind='barh', color=['steelblue', 'salmon'])

plt.title('Perbandingan Rata-rata Variabel\nData Regresi vs kab_kota_sulsel', fontsize=14)
plt.xlabel('Nilai Rata-rata', fontsize=12)
plt.ylabel('Variabel', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #73ad20e0; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;">  2. Radar Chart (Spider Plot) untuk Perbandingan Profil Variabel
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

#Mengambil data
labels = comparison_df.index.tolist()
values1 = comparison_df['Rata-rata Data_Regresi'].fillna(0).tolist()
values2 = comparison_df['Rata-rata kab_kota_sulsel'].fillna(0).tolist()

#Setup sudut
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values1 += values1[:1]
values2 += values2[:1]
angles += angles[:1]

#Plot
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values1, label='Data_Regresi', color='steelblue')
ax.fill(angles, values1, alpha=0.25, color='steelblue')
ax.plot(angles, values2, label='kab_kota_sulsel', color='salmon')
ax.fill(angles, values2, alpha=0.25, color='salmon')

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.title('Perbandingan Profil Rata-rata Variabel', fontsize=14)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

<div style="font-family: Arial; font-size: 20px; border-radius: 5px; background-color: #73ad20e0; padding: 2px; line-height: 2.0; color: #FFFFFF; text-align: center;"> 3. Line Plot Perbandingan Rata-rata
</div>

```python
plt.figure(figsize=(8, 5))
plt.plot(comparison_df.index, comparison_df['Rata-rata Data_Regresi'], marker='o', label='Data_Regresi', color='steelblue')
plt.plot(comparison_df.index, comparison_df['Rata-rata kab_kota_sulsel'], marker='o', label='kab_kota_sulsel', color='salmon')
plt.title("Perbandingan Rata-rata Variabel", fontsize=14)
plt.xlabel("Variabel")
plt.ylabel("Nilai Rata-rata")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
