```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("data_penjualan.csv")

df.info()
```

```python
sns.boxplot(x=df['Jenis Barang'])
```

```python
sns.boxplot(y=df['Jumlah'])
```

```python
sns.boxplot(data=df, x="Jenis Barang", y='Jumlah', color="grey")
```

```python
custom_pal = {"Elektronik": "#FF32C1", "Pakaian": "yellow", "Kosmetik":"g"}

sns.boxplot(data=df, x="Jenis Barang", y='Jumlah', palette=custom_pal,
  linewidth=0.22,
            width=0.12)
```

```python
custom_pal = {"Elektronik": "#FF32C1", "Pakaian": "yellow", "Kosmetik":"g"}

sns.boxplot(data=df, x="Jenis Barang", y='Jumlah', palette=custom_pal,
  linewidth=0.22,
            width=0.12,
notch=True,
fliersize=10,
flierprops={"marker": "o"})
```

```python
#Mengatur ukuran frame
plt.figure(figsize=(8, 6))
sns.boxplot(data=df,

            x="Jenis Barang", y='Jumlah', hue="Harga Satuan",

            palette="bright",

            linewidth=0.8,

            width=0.5,

            fliersize=3,

            flierprops={"marker": "o"})
plt.legend(bbox_to_anchor=(1, 0.6), loc=2, borderaxespad=1)
```

```python
#Menyiapkan frame 1 baris dan 2 kolom (untuk 2 diagram)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=False)

custom_pal = {"Elektronik": "#0C6F0D", "Pakaian": "#1DA11F", "Kosmetik":"#47EB4A"}
sns.boxplot(data=df, x="Jenis Barang", y='Jumlah', width=0.6, ax=axes[0],  palette=custom_pal,)
sns.boxplot(data=df, x="Jenis Barang", y='Jumlah', width=0.6, ax=axes[1],  palette=custom_pal,)

fig.tight_layout()
```

```python
# menggabungkan scatter plot dengan tren line
import numpy as np
import matplotlib.pyplot as plt

# Data
x = [67, -10, 32, 55, 100]
y = [21, -17, 42, -13, 77]

# ubah jadi array numpy
x = np.array(x)
y = np.array(y)

# Scatter plot titik-titik yang benar-benar tersebar
plt.scatter(x, y, color='blue', s=50, label='Data')

# Hitung trend line (regresi linear)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label='Trend Line')

plt.title("Scatter Plot dengan Tren Line")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend()
plt.show()
```
